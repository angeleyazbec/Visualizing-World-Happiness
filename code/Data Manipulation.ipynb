```python
import pandas as pd
```

```python
#read in csv
happiness_df = pd.read_csv("code/happiness.csv", index_col=0)

happiness_df
```

```python
#filter out years outside of 2018-2021
interest_years = happiness_df[happiness_df['Year'] > 2017]
interest_years
```

```python
# removing any n/a
clean_happiness = interest_years.dropna(axis=0, how="any")
clean_happiness
```

```python
#filter data for pre-COVID
pre_covid = clean_happiness[clean_happiness['Year'] < 2020]
pre_covid
```

```python
#filter data for during/post-COVID
post_covid = clean_happiness[clean_happiness['Year'] >= 2020]
post_covid
```

```python
#obtain countries list for years 2018-2019
pre_countries = pre_covid['Country']
pre_countries
```

```python
#obtain countries list for years 2020-2021
post_countries = post_covid['Country']
post_countries
```

```python
# using intersection to obtain common elements in a list
countries_list = (set(post_countries).intersection(set(pre_countries)))
countries_list
```

```python
# using isin function to filter out countries that are not of interest
# verifying the isin function captures only the countries that are desired
filtered_out = interest_years[~interest_years['Country'].isin(countries_list)]
filtered_out
```

```python
# using isin function to retain only the countries of interest
# as ascertained by the countries_list
filtered = interest_years[interest_years['Country'].isin(countries_list)]
filtered
```

```python
#read data to csv
filtered.to_csv("happiness_years.csv")
```
