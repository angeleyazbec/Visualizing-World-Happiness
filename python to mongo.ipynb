```python
# Module used to connect Python with MongoDb
import pymongo
import pandas as pd
import glob
import json
```

```python
#using glob to obtain the list of files
folder_path = 'Resources'
file_list = glob.glob(folder_path + "/*.csv")
file_list
```

```python
# reading in multiple files using Map
li_mapper = map(lambda filename: pd.read_csv(filename, index_col= None, header=None), file_list)
li_2 = list(li_mapper)
df = pd.concat(li_2, axis=0, ignore_index=True)
df
```

```python
#renaming all columns
df.columns = ['title', 'author', 'ATOS', 'NF/F', 'AR_book_link']
df
```

```python
#identifying duplicate book titles
dup_book = df.loc[df.duplicated(['title'], keep= "last")]
dup_book
```

```python
#dropping dpulicate titles
cleaned_wkar= df.drop_duplicates(subset=['title'], keep='last', inplace=False, ignore_index=False)
cleaned_wkar
```

```python
#writing the file to a csv to future use
cleaned_wkar.to_csv("Resources/clean_books.csv", index=False, header=True)
```

```python
#converting to a JSON for future use
records = json.loads(cleaned_wkar.to_json(orient='records'))

records
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the 'kid_books' database in Mongo
db = client.wkar
```

```python
db.collection.insert_many(cleaned_wkar.to_dict('records'))
```
